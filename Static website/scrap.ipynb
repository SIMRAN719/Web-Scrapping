```python
comments=soup.find_all(lambda text: isinstance(text,Comment))
for i in comments:
    print(i)
print(comments)
```

```python
pnames=soup.find_all(class_='ds-text-tight-l')
print(pnames)
```

```python
for names in soup.find_all(class_='ds-text-tight-l'):
    name=names.get_text()
    print(name)
```

```python
for age in soup.find_all(class_='ds-text-tight-m ds-font-regular ds-text-typo-mid3'):
    print(age.get_text())
```

```python
for i in soup.find_all(class_='ds-text-tight-m ds-font-regular ds-text-typo-mid3'):
    age=i.get_text()[5:7]
    print(age)
```

```python
data=[]
names=soup.find_all(class_='ds-text-tight-l')
ages=soup.find_all(class_='ds-text-tight-m ds-font-regular ds-text-typo-mid3')
for name,age in zip(names,ages):
    temp={}
    x=name.get_text()
    y=age.get_text()[5:7]
    temp['Name']=x
    temp['Age']=int(y)
    data.append(temp)
print(data)
```

```python
csv_file='data.csv'
col_titles=data[0].keys()
f=open(csv_file,'w',newline='')
writer=csv.DictWriter(f,fieldnames=col_titles)
writer.writeheader()
for row in data:
    writer.writerow(row)
```
